```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from pylab import *
%matplotlib inline

purchases = np.random.normal(3.0, 1.0, 1000)
credit = 25 + (purchases + np.random.normal(0, 0.5, 1000)) * 3
```

```python
scatter(purchases, credit)
```

```python
np.corrcoef(purchases, credit)
```

```python
slope, intersept, _, _, _ = stats.linregress(purchases, credit)
```

```python
def predict(x):
    return slope*x+intersept
```

```python
fitLine = predict(purchases)
plt.scatter(purchases, credit)
plt.plot(purchases, fitLine, c='r')
plt.show()
```

```python
purchases_ = 7
print('    ', purchases_, ': ', predict(purchases_))
```

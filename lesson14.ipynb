```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
incomes_1 = np.random.normal(27000, 15000, 10000)
incomes_2 = np.concatenate((incomes_1, np.random.normal(150000, 150, 100)))
```

```python
plt.hist(incomes_2, 50)
plt.show()
```

```python
plt.boxplot(incomes_1)
plt.show()
```

```python
plt.boxplot(incomes_2)
plt.show()
```

```python
print(np.std(incomes_1))
print(np.std(incomes_2))
```

```python
print(np.var(incomes_1))
print(np.var(incomes_2))
```

```python
import scipy.stats as sp

print(sp.skew(incomes_1))
print(sp.skew(incomes_2))
```

```python
print(sp.kurtosis(incomes_1))
print(sp.kurtosis(incomes_2))
```

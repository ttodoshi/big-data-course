```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-4, 4, 0.01)
plt.plot(x, x**2)
plt.plot(x, x**3)
```

```python
from random import randint
values = []
for _ in range(5):
    values.append(randint(0, 50))
colors = ['r', 'g', 'b', 'c', 'm']
labels = ['1', '2', '3', '4', '5']
plt.title('Кружок')
plt.pie(values, colors = colors, labels = labels)
```
